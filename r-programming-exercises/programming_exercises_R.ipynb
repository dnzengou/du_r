### Trends in scatterplots

Before we have a look at time series data, let’s return to an example that we studied in the last computer exercise;<br>
**the relationship between animal brain size and sleep times**:



```
ggplot(msleep, aes(brainwt, sleep_total)) +
geom_point() +
xlab("Brain weight (logarithmic scale)") +
ylab("Total sleep time") +
scale_x_log10()
```



There appears to be a decreasing trend in the plot. To aid the eye, we can add a smoothed line by adding a new geom, geom_smooth, to the figure:



```
ggplot(msleep, aes(brainwt, sleep_total)) +
geom_point() +
geom_smooth() +
xlab("Brain weight (logarithmic scale)") +
ylab("Total sleep time") +
scale_x_log10()
```

Check the documentation for geom_smooth. Starting with the plot of brain weight vs. sleep time created

this case, more or less smoothing?

````r
ggplot(msleep, aes(brainwt, sleep_total)) +
geom_point() + 
geom_smooth(span=0.3) +
xlab("Brain weight (log scale)") +
ylab("total sleep time") +
scale_x_log10()
````

````r
?geom_smooth
````

````r
ggplot(msleep, aes(brainwt, sleep_total)) +
geom_point() + 
# see doc. geom_smooth to remove the confidence intervalle (grey shade)
# around smooth lines, by default TRUE
geom_smooth(span=0.3, se = FALSE) +
xlab("Brain weight (log scale)") +
ylab("total sleep time") +
scale_x_log10()
````

````r
## OR change the confidence level (95% default value) to lower  
ggplot(msleep, aes(brainwt, sleep_total)) +
geom_point() + 
geom_smooth(span=0.3, level = 0.4) +
xlab("Brain weight (log scale)") +
ylab("total sleep time") +
scale_x_log10()
````

````r
## OR change the confidence level (95% default value) to lower  
ggplot(msleep, aes(brainwt, sleep_total, colour = "red")) +
#ggplot(msleep, aes(brainwt, sleep_total, color=family)) +
geom_point() + 
geom_smooth(span=0.3, level = 0.4) +
xlab("Brain weight (log scale)") +
ylab("total sleep time") +
scale_x_log10()
````
